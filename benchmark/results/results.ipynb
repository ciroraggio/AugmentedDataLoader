```python
import pandas as pd
test_res = pd.read_csv("./results.csv")
test_res
```

```python
import matplotlib.pyplot as plt

# Dati
test = test_res["test"].values
cpu = test_res["cpu"].values
gpu = test_res["gpu"].values
transf = test_res["transf"].values

plt.plot(cpu, transf, color='blue', label='CPU')
plt.plot(gpu, transf, color='green', label='GPU')
plt.scatter(cpu, transf, color='blue')
plt.scatter(gpu, transf, color='green')

plt.xlabel('Tempo [s]')
plt.ylabel('Numero di trasformazioni')
plt.title('Tempo di esecuzione al variare delle trasformazioni (CPU vs GPU)')
plt.legend(loc='lower right')
plt.show()
```

```python
import matplotlib.pyplot as plt

test = test_res["test"].values
cpu = test_res["cpu"].values
gpu = test_res["gpu"].values
transf = test_res["transf"].values

bar_width = 0.35

x = transf  # Utilizza i valori di "transf" come posizioni sull'asse x

plt.bar(x, cpu, width=bar_width, color='blue', label='CPU')

# Sposta le barre del gruppo GPU a destra
x_gpu = [pos + bar_width for pos in x]
plt.bar(x_gpu, gpu, width=bar_width, color='green', label='GPU')

plt.xlabel('Numero di trasformazioni')
plt.ylabel('Tempo [s]')
plt.title('Tempo di esecuzione al variare delle trasformazioni (CPU vs GPU)')
plt.legend()
plt.show()
```

```python
import matplotlib.pyplot as plt

test = test_res["test"].values
cpu = test_res["cpu"].values
gpu = test_res["gpu"].values
transf = test_res["transf"].values

bar_height = 0.35  # Altezza delle barre

y = transf  # Utilizza i valori di "transf" come posizioni sull'asse y

plt.barh(y, cpu, height=bar_height, color='blue', label='CPU')

# Sposta le barre del gruppo GPU in basso
y_gpu = [pos + bar_height for pos in y]
plt.barh(y_gpu, gpu, height=bar_height, color='green', label='GPU')

plt.ylabel('Numero di trasformazioni')
plt.xlabel('Tempo [s]')
plt.title('Tempo di esecuzione al variare delle trasformazioni (CPU vs GPU)')
plt.legend()
plt.show()
```
